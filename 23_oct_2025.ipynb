```python
# Fill numeric columns with their mean
df1_filled = df1.copy()

for col in df1_filled.columns:
    if df1_filled[col].dtype in ["float64", "int64"]:
        df1_filled[col].fillna(df1_filled[col].mean(), inplace=True)
    else:
        # For text/categorical data (like Sample_ID), fill with mode
        df1_filled[col].fillna(df1_filled[col].mode()[0], inplace=True)

print(df1_filled)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)  # reproducible

# Define columns
columns = [
    "Sample_ID",
    "Iron_Class",
    "Zinc_Class",
    "pH_Level_Class",
    "Lactic_Acid_Class"
]

# Class options
classes = ["Low", "Medium", "High"]

# Create data
data = {
    "Sample_ID": [f"S{i+1}" for i in range(10)]
}

# Fill the remaining 4 columns with random classes
for col in columns[1:]:
    data[col] = np.random.choice(classes, size=10)

# Create DataFrame
df_small_class = pd.DataFrame(data)

# Introduce 20% NaN values randomly (only in class columns)
mask = np.random.rand(*df_small_class.iloc[:, 1:].shape) < 0.2
df_small_class.iloc[:, 1:] = df_small_class.iloc[:, 1:].mask(mask)

# Show dataset with empty values
print(df_small_class)
```

```python
# Make a copy to preserve original data
df_filled = df_small_class.copy()

# Fill NaN in each column with the most frequent class (mode)
for col in df_filled.columns[1:]:  # skip Sample_ID
    df_filled[col].fillna(df_filled[col].mode()[0], inplace=True)

# Show the filled dataset
print(df_filled)
```

```python
# Mean
mean_val = df1["Iron_mg_L"].mean()
print("Mean:", mean_val)

# Median
median_val = df1["Iron_mg_L"].median()
print("Median:", median_val)

# Mode (most frequent value)
mode_val = df1["Iron_mg_L"].mode()[0]  # [0] to get the value from the series
print("Mode:", mode_val)

# Average (same as mean)
average_val = df1["Iron_mg_L"].mean()
print("Average:", average_val)
```

```python
# Maximum value
max_val = df1["Iron_mg_L"].max()
print("Maximum:", max_val)

# Minimum value
min_val = df1["Iron_mg_L"].min()
print("Minimum:", min_val)

# Total / Sum of all values
total_val = df1["Iron_mg_L"].sum()
print("Total:", total_val)
```
