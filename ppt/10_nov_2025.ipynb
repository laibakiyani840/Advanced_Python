#### **Create a Python tool that computes the roots of a quadratic equation using functions and conditional logic. Plot the corresponding parabola using matplotlib and display the nature of the roots.**

```python
"""
Quadratic Equation Application
Main application to use the quadratic analyzer tool
"""

from quadratic_tool import QuadraticAnalyzer, quick_analyze, solve_quadratic
import matplotlib.pyplot as plt

def demonstrate_examples():
    """Demonstrate the tool with example equations"""
    
    print("QUADRATIC EQUATION ANALYZER - DEMONSTRATION")
    print("=" * 60)
    
    examples = [
        (1, -5, 6),    # Real distinct roots: x - 5x + 6 = 0
        (1, -4, 4),    # Real equal roots: x - 4x + 4 = 0
        (1, 2, 5),     # Complex roots: x + 2x + 5 = 0
        (2, 3, -2),    # Real distinct: 2x + 3x - 2 = 0
        (-1, 0, 4)     # Real distinct: -x + 4 = 0
    ]
    
    for i, (a, b, c) in enumerate(examples, 1):
        print(f"\nExample {i}:")
        try:
            analyzer = QuadraticAnalyzer(a, b, c)
            analyzer.display_analysis()
            
            # Ask if user wants to plot
            plot_choice = input("Plot graph? (y/n): ").lower()
            if plot_choice == 'y':
                analyzer.plot_parabola()
                
        except Exception as e:
            print(f"Error: {e}")

def interactive_mode():
    """Interactive mode for user input"""
    
    print("\n" + "=" * 60)
    print("INTERACTIVE QUADRATIC SOLVER")
    print("=" * 60)
    
    while True:
        try:
            print("\nEnter coefficients for quadratic equation ax + bx + c = 0")
            a = float(input("Enter coefficient a: "))
            b = float(input("Enter coefficient b: "))
            c = float(input("Enter coefficient c: "))
            
            # Analyze the equation
            analyzer = QuadraticAnalyzer(a, b, c)
            analyzer.display_analysis()
            
            # Ask for plot
            plot_choice = input("\nDo you want to plot the parabola? (y/n): ").lower()
            if plot_choice == 'y':
                analyzer.plot_parabola()
            
            # Continue or exit
            continue_choice = input("\nAnalyze another equation? (y/n): ").lower()
            if continue_choice != 'y':
                break
                
        except ValueError:
            print("Error: Please enter valid numbers!")
        except Exception as e:
            print(f"Error: {e}")

def quick_solve_mode():
    """Quick solving mode without detailed analysis"""
    
    print("\n" + "=" * 60)
    print("QUICK SOLVE MODE")
    print("=" * 60)
    
    try:
        print("Enter coefficients for quick solution:")
        a = float(input("a: "))
        b = float(input("b: "))
        c = float(input("c: "))
        
        nature, roots, discriminant = solve_quadratic(a, b, c)
        
        print(f"\nQuick Results:")
        print(f"Equation: {a}x + {b}x + {c} = 0")
        print(f"Discriminant: {discriminant:.2f}")
        print(f"Nature of Roots: {nature}")
        print(f"Roots: {roots}")
        
    except Exception as e:
        print(f"Error: {e}")

def main():
    """Main application function"""
    
    print(" QUADRATIC EQUATION ANALYZER TOOL")
    print("A comprehensive tool for solving and visualizing quadratic equations")
    
    while True:
        print("\n" + "=" * 60)
        print("MAIN MENU")
        print("=" * 60)
        print("1. View Demonstration Examples")
        print("2. Interactive Equation Solver")
        print("3. Quick Solve Mode")
        print("4. Exit")
        
        choice = input("\nEnter your choice (1-4): ").strip()
        
        if choice == '1':
            demonstrate_examples()
        elif choice == '2':
            interactive_mode()
        elif choice == '3':
            quick_solve_mode()
        elif choice == '4':
            print("\nThank you for using Quadratic Equation Analyzer! ")
            break
        else:
            print("Invalid choice! Please enter 1-4.")

if __name__ == "__main__":
    main()
```
