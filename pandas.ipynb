```python
import pandas as pd
print(pd.__version__)
```

## Default index

```python
s = pd.Series([10, 20, 30, 40, 50])
print(s)
```

## Manual Index

```python
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
print(s)
```

## Standard

```python
data = [10, 20, 30]
labels = ['b', 'c', 'a']
series = pd.Series(data, index=labels)
print(series)
```

## Sorting index

```python
series = series.sort_index()
print(series)
```

## Sorting Values

```python
series = series.sort_values()
print(series)
```

## Accessing Data

```python
print(series['a'])
print(series[['a', 'c']])
```

```python
import pandas as pd

counts = pd.Series({'E_coli': 120, 'S_aureus': 45, 'B_subtilis': 10, 'L_lactis': 25})
print("Counts:\n", counts)
print("\nTop species:\n", counts.sort_values(ascending=False).head(3))
```
